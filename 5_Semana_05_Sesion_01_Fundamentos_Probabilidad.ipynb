## Introducción a la Probabilidad: Del Concepto a la Práctica en Ingeniería

### 1. Contexto en Ingeniería
En las diversas ramas de la ingeniería (civil, agrícola, agroindustrial), la incertidumbre es una constante. Ya sea al predecir la resistencia de un material, estimar el rendimiento de un cultivo bajo condiciones climáticas variables, o evaluar la calidad de un lote de producción, la probabilidad se convierte en nuestro lenguaje matemático fundamental para cuantificar y modelar esta incertidumbre, permitiéndonos tomar decisiones informadas y gestionar riesgos.

### 2. Conceptos Fundamentales
Para abordar la probabilidad, es esencial comprender sus pilares:

*   **Experimento Aleatorio:** Cualquier proceso que puede repetirse bajo las mismas condiciones y cuyos resultados no pueden predecirse con certeza. Por ejemplo, la inspección de una viga de concreto o el lanzamiento de un dado.
*   **Espacio Muestral ($\Omega$):** El conjunto de todos los posibles resultados de un experimento aleatorio. Si inspeccionamos una viga, $\Omega = \{$'Aprobado', 'Fallo_Menor', 'Fallo_Critico'$\}$.
*   **Evento ($A$):** Cualquier subconjunto del espacio muestral. Un evento podría ser 'La viga tiene un Fallo Crítico'.

### 3. Ecuaciones de Probabilidad

La probabilidad de que ocurra un evento puede ser abordada desde dos perspectivas principales:

*   **Probabilidad Teórica (Regla de Laplace):** Si todos los resultados en un espacio muestral finito son igualmente probables, la probabilidad de un evento $A$ se calcula como:
    $$
P(A) = \frac{\text{Número de resultados favorables a } A}{\text{Número total de resultados posibles}}
    $$

*   **Probabilidad Empírica (Frecuencia Relativa):** Basada en la observación de experimentos repetidos. Si un experimento se realiza $n$ veces y un evento $A$ ocurre $k$ veces, la probabilidad empírica de $A$ es:
    $$
P(A) \approx \frac{\text{Número de veces que ocurre } A}{\text{Número total de ensayos}}
    $$

### 4. Transición Práctica
A continuación, utilizaremos Python para simular un experimento aleatorio y calcular estas probabilidades empíricas a partir de datos, conectando así la teoría abstracta con la realidad de la ingeniería.

## **Celda 3: Código Python (Espacio Muestral, Eventos y Probabilidad Empírica)**

````
import pandas as pd
import numpy as np

# Simulamos el registro histórico de inspección de 500 vigas de concreto (Ing. Civil)
# o lotes de producción (Ing. Agroindustrial).
# Resultados posibles (Espacio Muestral): "Aprobado", "Fallo_Menor", "Fallo_Critico"

np.random.seed(42)
espacio_muestral = ['Aprobado', 'Fallo_Menor', 'Fallo_Critico']
# Asignamos una probabilidad teórica subyacente (80% éxito, 15% fallo menor, 5% fallo crítico)
inspecciones = np.random.choice(espacio_muestral, size=500, p=[0.80, 0.15, 0.05])

df_calidad = pd.DataFrame({'Estado_Inspeccion': inspecciones})

# ==========================================
# CÁLCULO DE PROBABILIDAD EMPÍRICA
# ==========================================
print("--- FRECUENCIAS ABSOLUTAS ---")
frecuencias = df_calidad['Estado_Inspeccion'].value_counts()
print(frecuencias)

print("\n--- PROBABILIDAD EMPÍRICA (Frecuencia Relativa) ---")
prob_empirica = df_calidad['Estado_Inspeccion'].value_counts(normalize=True)
print(prob_empirica.round(4))

# Interpretación Ingenieril:
# El "Espacio Muestral" son los 3 estados posibles.
# Un "Evento" puede ser que una viga tenga un 'Fallo_Critico'.
# La Probabilidad Empírica de 'Fallo_Critico' calculada con nuestros datos es cercana a 0.05 (5%),
# calculada simplemente dividiendo los casos favorables entre el total de inspecciones.
````

## Probabilidad Condicional e Independencia: Profundizando en las Relaciones

### 1. Introducción Práctica
Hasta ahora, hemos calculado probabilidades empíricas de eventos individuales. Sin embargo, en ingeniería, a menudo nos enfrentamos a situaciones donde la ocurrencia de un evento puede influir en la probabilidad de otro. Por ejemplo, ¿es la probabilidad de un fallo crítico en una viga o en un lote de producción la misma, independientemente del proveedor que suministró el material? Comprender esta interdependencia es crucial para el control de calidad y la toma de decisiones.

### 2. Conceptos Clave

*   **Probabilidad Condicional:** Es la probabilidad de que un evento $A$ ocurra, *dado* que otro evento $B$ ya ha ocurrido. Actualiza nuestra medida de incertidumbre sobre $A$ basándonos en nueva información ($B$).

*   **Independencia Estadística:** Dos eventos $A$ y $B$ son estadísticamente independientes si la ocurrencia de uno no afecta la probabilidad de ocurrencia del otro. En un contexto de ingeniería, esto significaría, por ejemplo, que la calidad de un material no depende de qué proveedor lo entregó, lo cual es un escenario ideal para el control de calidad.

### 3. Ecuaciones Fundamentales

*   La **Probabilidad Condicional** de $A$ dado $B$ se define como:
    $$
P(A|B) = \frac{P(A \cap B)}{P(B)}
    $$
    donde $P(B) > 0$. $P(A \cap B)$ es la probabilidad de que $A$ y $B$ ocurran simultáneamente.

*   Dos eventos $A$ y $B$ son **Independientes** si se cumple cualquiera de las siguientes condiciones:
    $$
P(A|B) = P(A)
    $$
    o, equivalentemente, si:
    $$
P(A \cap B) = P(A) \cdot P(B)
    $$

### 4. Transición al Código: Análisis con Tablas de Contingencia
A continuación, utilizaremos la librería Pandas de Python para construir una tabla de contingencia cruzada. Esto nos permitirá visualizar la relación entre el estado de inspección y el proveedor, y calcular las probabilidades condicionales directamente a partir de nuestros datos simulados, evaluando si existe independencia entre estas variables.

## **Celda 4: Código Python (Probabilidad Condicional e Independencia)**

````
# Ahora, supongamos que estas inspecciones provienen de dos proveedores distintos
proveedores = np.random.choice(['Proveedor_A', 'Proveedor_B'], size=500, p=[0.6, 0.4])
df_calidad['Proveedor'] = proveedores

# Tabla de contingencia cruzada
tabla_cruzada = pd.crosstab(df_calidad['Proveedor'], df_calidad['Estado_Inspeccion'], margins=True)

print("--- TABLA DE CONTINGENCIA ---")
print(tabla_cruzada)

# Cálculo de Probabilidad Condicional
# P(Fallo_Critico | Proveedor_B) = Casos de Fallo Crítico en Prov_B / Total de Prov_B
total_prov_b = tabla_cruzada.loc['Proveedor_B', 'All']
fallos_prov_b = tabla_cruzada.loc['Proveedor_B', 'Fallo_Critico']

prob_condicional = fallos_prov_b / total_prov_b
print(f"\nProbabilidad de Fallo Crítico DADO que es del Proveedor B: {prob_condicional:.4f}")

# Interpretación: Si la probabilidad condicional es igual a la probabilidad marginal (general),
# decimos que los eventos son INDEPENDIENTES. En la práctica ingenieril, si un proveedor
# tiene una tasa de fallos mucho mayor, la calidad "depende" del proveedor.
````
