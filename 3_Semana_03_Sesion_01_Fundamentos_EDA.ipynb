```markdown
### Explicación del Código para la Detección de Valores Faltantes

Este fragmento de código se enfoca en identificar y resumir la presencia de valores faltantes (NaN) en el DataFrame `datos_ingenieria`, lo cual es el segundo paso en un proceso de Análisis Exploratorio de Datos (EDA) estructurado.

```python
# ==========================================
# PASO 2: VALORES FALTANTES
# ==========================================
print("\n--- 2. DETECCIÓN DE VALORES FALTANTES ---")
# Imprime un encabezado para indicar el inicio del Paso 2 del EDA, que es la detección de valores faltantes.

faltantes = datos_ingenieria.isna().sum()
# `datos_ingenieria.isna()`: Crea un DataFrame booleano del mismo tamaño que `datos_ingenieria`, donde `True` indica un valor faltante (NaN) y `False` un valor presente.
# `.sum()`: Suma los valores `True` (que se interpretan como 1) por cada columna. Esto cuenta el número total de valores faltantes en cada columna.
# El resultado se almacena en la variable `faltantes`, que será una Serie de Pandas donde el índice son los nombres de las columnas y los valores son los conteos de NaN.

print(faltantes[faltantes > 0])
# Filtra la Serie `faltantes` para mostrar solo aquellas columnas que tienen al menos un valor faltante (es decir, cuyo conteo es mayor que 0).
# Esto permite al usuario ver rápidamente qué columnas tienen NaN y cuántos tienen, sin mostrar las columnas que están completas.

# Interpretación: Saber cuántos faltan y en qué variables nos dice si debemos
# imputar datos o descartar registros antes de calcular estadísticas.
# Este comentario explica la importancia ingenieril y práctica de este paso: la cantidad y ubicación de los datos faltantes influye en las decisiones de preprocesamiento, como la imputación (rellenar los NaN) o la eliminación de filas/columnas.
```


#### **Celda 5: Código Python (Paso 4)**

````
# ==========================================
# PASO 4: FORMA DE LA DISTRIBUCIÓN
# ==========================================
print("\n--- 4. ASIMETRÍA (SKEWNESS) Y CURTOSIS ---")

# Calculamos asimetría y curtosis solo para variables numéricas, omitiendo los NaN
resistencia_limpia = datos_ingenieria['Resistencia_MPa'].dropna()

asimetria = skew(resistencia_limpia)
curtosis_val = kurtosis(resistencia_limpia)

print(f"Asimetría de la Resistencia: {asimetria:.2f}")
print(f"Curtosis de la Resistencia: {curtosis_val:.2f}")

# Interpretación Ingenieril:
# - Asimetría cercana a 0 indica distribución simétrica.
# - Curtosis > 0 indica colas pesadas (mayor probabilidad de valores extremos o fallos inesperados en el material).
````

```markdown
### Explicación del Código para el Análisis de la Forma de la Distribución (Asimetría y Curtosis)

Este fragmento de código calcula la asimetría (skewness) y la curtosis de la variable `Resistencia_MPa`, lo cual es el cuarto paso en un proceso de Análisis Exploratorio de Datos (EDA) estructurado. Estas métricas ayudan a entender la forma de la distribución de los datos.

```python
# ==========================================
# PASO 4: FORMA DE LA DISTRIBUCIÓN
# ==========================================
print("\n--- 4. ASIMETRÍA (SKEWNESS) Y CURTOSIS ---")
# Imprime un encabezado para indicar el inicio del Paso 4 del EDA, que es el análisis de la forma de la distribución.

# Calculamos asimetría y curtosis solo para variables numéricas, omitiendo los NaN
resistencia_limpia = datos_ingenieria['Resistencia_MPa'].dropna()
# Selecciona la columna 'Resistencia_MPa' del DataFrame `datos_ingenieria`.
# `.dropna()`: Elimina cualquier valor faltante (NaN) de esta Serie. Es crucial para que los cálculos de asimetría y curtosis sean precisos y no se vean afectados por los NaN.
# El resultado (una Serie sin NaN) se almacena en `resistencia_limpia`.

asimetria = skew(resistencia_limpia)
# Calcula la asimetría (skewness) de la Serie `resistencia_limpia` utilizando la función `skew` importada de `scipy.stats`.
# La asimetría mide la simetría de la distribución. Un valor de 0 indica simetría perfecta.

curtosis_val = kurtosis(resistencia_limpia)
# Calcula la curtosis de la Serie `resistencia_limpia` utilizando la función `kurtosis` importada de `scipy.stats`.
# La curtosis mide la "puntiagudez" de la distribución y la pesadez de sus colas (la probabilidad de valores extremos).

print(f"Asimetría de la Resistencia: {asimetria:.2f}")
# Imprime el valor calculado de la asimetría, formateado a dos decimales.

print(f"Curtosis de la Resistencia: {curtosis_val:.2f}")
# Imprime el valor calculado de la curtosis, formateado a dos decimales.

# Interpretación Ingenieril:
# - Asimetría cercana a 0 indica distribución simétrica.
# - Curtosis > 0 indica colas pesadas (mayor probabilidad de valores extremos o fallos inesperados en el material).
# Estos comentarios proporcionan el contexto ingenieril y la interpretación práctica de los valores de asimetría y curtosis en el dominio de la ingeniería de materiales.
```
